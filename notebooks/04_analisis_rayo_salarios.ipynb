# Análisis salarial del Rayo Vallecano

Este notebook carga el CSV y genera 10 estadísticas relevantes:
1. Jugadores que terminan contrato antes.
2. Top salarios anuales.
3. Top salarios semanales.
4. Reservas más caras.
5. Titulares más caros.
6. Masa salarial por rol.
7. Edad media por rol.
8. Jugadores con mayor compromiso económico restante.
9. Distribución salarial por posición.
10. Cláusulas de rescisión más altas.


```python

import pandas as pd

df = pd.read_csv(r"rayo_vallecano_salarios.csv")
df.head()
```

## 1. Jugadores que terminan contrato antes

```python

df.sort_values('Años restantes')[['Jugador','Vencimiento','Años restantes','Salario fijo anual (EUR)']].head(20)
```

## 2. Top 10 salarios anuales

```python

df.sort_values('Total anual (EUR)', ascending=False)[['Jugador','Salario fijo anual (EUR)','Total anual (EUR)','Rol']].head(10)
```

## 3. Top 10 salarios semanales

```python

df.sort_values('Salario semanal (EUR)', ascending=False)[['Jugador','Salario semanal (EUR)']].head(10)
```

## 4. Reservas más caras

```python

df[df['Rol']=='Reserve'].sort_values('Total anual (EUR)', ascending=False)[['Jugador','Total anual (EUR)']].head(10)
```

## 5. Titulares más caros

```python

df[df['Rol']=='Starter'].sort_values('Total anual (EUR)', ascending=False)[['Jugador','Total anual (EUR)']].head(10)
```

## 6. Masa salarial por rol

```python

df.groupby('Rol')['Total anual (EUR)'].agg(['count','sum','mean']).sort_values('sum', ascending=False)
```

## 7. Edad media por rol

```python

df.groupby('Rol')['Edad'].agg(['mean','min','max'])
```

## 8. Mayor compromiso económico restante

```python

df.sort_values('Importe restante (EUR)', ascending=False)[['Jugador','Importe restante (EUR)','Años restantes']].head(10)
```

## 9. Salario medio por posición

```python

df.groupby('Posición')['Total anual (EUR)'].mean().sort_values(ascending=False)
```

## 10. Cláusulas de rescisión más altas

```python

df[['Jugador','Cláusula rescisión (EUR)']].dropna().sort_values('Cláusula rescisión (EUR)', ascending=False).head(10)
```
